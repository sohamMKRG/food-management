# **Problem Statement**


Food wastage is a significant issue, with many households and restaurants discarding surplus food while numerous people struggle with food insecurity. This project aims to develop a Local Food Wastage Management System, where:

Restaurants and individuals can list surplus food.

NGOs or individuals in need can claim the food.

#### **Define Your Business Objective?**

The primary business objective of the Local Food Wastage Management System is to maximize the efficiency of surplus food redistribution to minimize waste and combat food insecurity.

This is achieved by establishing a direct and transparent connection between food providers and those in need. The system's success is defined by three key goals:

Reduce Food Waste: By providing a structured platform for providers (e.g., restaurants, grocery stores) to list their excess food, the system aims to divert a significant volume of edible food from landfills. This directly contributes to environmental sustainability and cost savings for businesses.

Increase Food Accessibility: The platform seeks to improve access to nutritious food for receivers (e.g., individuals, NGOs, shelters) by making available food listings easy to discover and claim. This directly supports social good by addressing food insecurity within local communities.

Provide Data-Driven Insights: A critical, long-term objective is to leverage the data collected to generate actionable insights. By analyzing trends in donations, claims, and logistics, the system can provide valuable reports to stakeholders. These insights can inform better decision-making, optimize distribution efforts, and ultimately, scale the system's impact.

In essence, the business objective is to create a self-sustaining ecosystem that transforms a societal problem—food wastage—into a productive process that benefits both businesses and the community.

### What did you know about your dataset?

Based on the provided CSV files and the project description, I have a detailed understanding of the dataset and its structure.

The dataset is composed of four separate but interconnected tables:

Providers: Contains information about the organizations or individuals donating food. I know their unique Provider_ID, Name, Type (e.g., Restaurant, Grocery Store), Address, City, and Contact information. The Provider_ID is a key identifier for linking this table to other parts of the database.

Receivers: Stores details about the individuals or organizations that claim the food. This table includes their unique Receiver_ID, Name, Type (e.g., NGO, Shelter), City, and Contact details. The Receiver_ID is a foreign key used in the claims table.

Food Listings: This is the central table for all available food items. For each listing, I have the unique Food_ID, Food_Name, Quantity, Expiry_Date, and the Provider_ID that links back to the provider table. It also contains categorical data like Provider_Type, Location, Food_Type (e.g., Vegetarian, Vegan), and Meal_Type (e.g., Breakfast, Dinner). This table is the primary source for the filtering and search functionality.

Claims: This table tracks the transactions. It connects a Food_ID with a Receiver_ID and records the Status of the claim (e.g., Completed, Pending, Canceled) and a Timestamp for when the claim was made. This table is essential for analyzing the system's operational efficiency and trends in food claims.

I understand that the dataset is "cleaned," meaning it's ready for direct use in the database without significant preprocessing. The relationships between these tables—specifically through Food_ID, Provider_ID, and Receiver_ID—form the basis for the complex SQL joins required for the project's data analysis and dashboard features.
